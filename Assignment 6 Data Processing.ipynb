### Part 1: Polarised Boundary Conditions

```python
import numpy as np
import os
import subprocess
from IPython.display import Image, display, clear_output
import matplotlib.pyplot as plt

# Set a larger font size for figures
plt.rc('font', size=15)

# This generates a list of data file names (plus paths) that the C program will write data to. It also creates a 
# directory for these files based on the experiment name.
def GenerateDataFileNames():
    DataDirectoryName = ExperimentName+'_data'
    if not os.path.exists(DataDirectoryName):
        os.mkdir(DataDirectoryName)
    DataFileNames = [f'{DataDirectoryName}/{DataDirectoryName}_Bh={Bh}_BJ={BJ}_S={S}_N={N}.dat' for N in NList]
    return DataDirectoryName, DataFileNames

# This calls the C program with appropriate parameters.
def GenerateData():
    CommandStrings = [f'Assignment6.exe {Bh} {BJ} {N} {S} {DataFileName} {1}' for N, DataFileName in zip(NList, DataFileNames)]
    for RunNumber, CommandString in enumerate(CommandStrings):
        clear_output(wait=True)
        display(f'Run {RunNumber+1} of {len(CommandStrings)}: '+ CommandString)
        subprocess.run(CommandString, shell=True, capture_output=True, text=True)
    clear_output(wait=True)
    display("Done!")
```

```python
# Used to name subdirectory where data will be stored.
ExperimentName = 'Assignment6Polarised'
# The number of sweeps to generate.
S = 20000
# A list of the lattice sizes we want to generate data for.
NList = [200]
# Other parameters
Bh = 0
BJ = 0.2
# The three variables declared above are used to generate the data directory and list of data file names.
DataDirectoryName, DataFileNames = GenerateDataFileNames()

# This can be commented out if the data is already generated and only needs to be read in the next line.
GenerateData()

# Read the data from the files generated by the C program.
AllData = np.array([np.genfromtxt(FileName, delimiter=',') for FileName in DataFileNames])
SweepData1 = np.zeros(S)
AbsSweepData1 = np.zeros(S)

RunningAverage1 = np.zeros(S)
AbsRunningAverage1 = np.zeros(S)

for i in range(S):
    SweepData1[i] = AllData[0,i,0]
    AbsSweepData1[i] = AllData[0,i,1]
    
    RunningAverage1[i] = np.sum(SweepData1)/(i+1)
    AbsRunningAverage1[i] = np.sum(AbsSweepData1)/(i+1)
```

```python
fig, (ax1)= plt.subplots(1, 1, figsize=(12, 6))

fig.tight_layout(pad=1)
time = np.arange(1,S+1,1,dtype=int)
y1 = SweepData1/(NList[0]*NList[0])
#y2 = AbsSweepData1/(NList[0]*NList[0])
y3 = RunningAverage1/(NList[0]*NList[0])
y4 = AbsRunningAverage1/(NList[0]*NList[0])
ax1.plot(time,y1,'o')
#ax1.plot(time,y2, 'o')
ax1.plot(time,y3, '-', linewidth = 2.5)
ax1.plot(time,y4, '-', linewidth = 2.5)
ax1.set(xlabel=r'Number of Sweeps', ylabel=r'Magnetisation', title=f'Plot for {S} sweeps on a polarised square lattice of side length L = {NList[0]}')
plt.ylim(-1.1,1.1)
plt.savefig(DataDirectoryName+'\L=200_BJ=0.2.png', dpi=300, transparent=False, bbox_inches='tight')
plt.show()
```
